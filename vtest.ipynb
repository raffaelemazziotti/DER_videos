```python
def to_python(obj):
    if isinstance(obj, dict):
        return {to_python(k): to_python(v) for k, v in obj.items()}
    elif isinstance(obj, (list, tuple)):
        return [to_python(x) for x in obj]
    elif hasattr(obj, "item"):   # NumPy scalar
        return obj.item()
    else:
        return obj


import json

all_info = {}

for vm in vmh:
    info = vm.info()
    info_py = to_python(info)  # convert numpy -> python
    all_info[vm.rec] = info_py

with open("video_info.json", "w") as f:
    json.dump(all_info, f, indent=4)
```
